```python
import numpy as np
```

```python
a = np.arange(0,100)
b = a - 5
c = a ** 3 - b ** 2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Создать  figure object
# Добавить axes к figure список [0, 0, 1, 1] list как параметр
# Создать график с данными по оси x и b по y оси
# Добавьте метки к графику, как на графике ниже
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(a, b)
axes.set_xlabel('a label')
axes.set_ylabel('b label')
axes.set_title('title label')
```

```python
#Создать объект фигуры и добавьте две оси - axes1 и axes2. 
#Передать следующие списки в качестве параметров - [0, 0, 1, 1] для осей1 
#и [0,2, 0,5, 0,2, 0,3] для осей2
figure = plt.figure()
axes1 = figure.add_axes([0, 0, 1, 1])
axes2 = figure.add_axes([0.2, 0.5, 0.2, 0.3])
```

```python
#Создайте графики с a в качестве данных оси x и b в качестве данных оси y на axes1 
#и с b в качестве данных оси x и c в качестве данных оси y на axes2 для предыдущего 
#объекта фигуры и покажите график, вызвав переменную фигуры
axes1.plot(a, b)
axes2.plot(b, c)
figure
```

```python
#Создайте объект фигуры и добавьте две оси - axes1 и axes2. 
#Передайте следующие списки в качестве параметров - [0, 0, 1, 1] для axes1 
#и [0,2, 0,5, 0,3, 0,4] для axes2
#Создайте графики, используя массивы a, b, c, чтобы они соответствовали изображению ниже. 
#Добавьте ограничения на графики - предел x равен (20, 60) и предел y равен (10, 70) для axes1 
#и предел x равен (20, 60) и предел y равен (10000, 70000) для axes2
figure = plt.figure()
axes1 = figure.add_axes([0, 0, 1, 1])
axes2 = figure.add_axes([0.2, 0.5, 0.3, 0.4])

axes1.plot(a, b)
axes1.set_xlabel('a array')
axes1.set_ylabel('b array')
axes1.set_title('Outer plot')
axes1.set_xlim(20, 60)
axes1.set_ylim(10, 70)

axes2.plot(a, c)
axes2.set_xlabel('a array')
axes2.set_ylabel('c array')
axes2.set_title('Inner plot')
axes2.set_xlim(20, 60)
axes2.set_ylim(10000, 70000)
```

```python
#Создать объект фигуры с 2 axes в 1 строке с помощью метода subplots()
#Изменить размер обеих axes - установитm 11 дюймов в ширину и 3 дюйма в высоту.
#Создать график на первых осях с c в качестве данных оси x и a в качестве данных оси y
#Создать график на второй оси с c в качестве данных оси x и b в качестве данных оси y
#Bзменить внешний вид сюжетных линий. Играйте с цветами, шириной и стилями
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(11, 3))
axes[0].plot(c, a, color='purple', ls='--', lw=2)
axes[1].plot(c, b, color='orange', ls=':', lw=5)
```
